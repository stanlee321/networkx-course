# Network Data Science with NetworkX and Python

## Add and remove nodes and edges from graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
graph = nx.random_internet_as_graph(20)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph.add_edge(18, 13)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph.add_node(40)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph.add_node('another_node')
graph.add_node(lambda x: x)

nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph.add_edge('A', 'B')
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.Graph()
new_nodes =['q', 'w', 'e', 'r', 't', 'y']
new_edges = [('q', 'w'), ('e', 'r'),  ('t', 'y'),
             ('q', 'y'), ('w', 't')]
graph.add_nodes_from(new_nodes)
graph.add_edges_from(new_edges)

nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```
