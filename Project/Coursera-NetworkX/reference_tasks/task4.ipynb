# Network Data Science with NetworkX and Python

## Generate random graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.complete_graph(10)
```

```python
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.binomial_tree(5)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.binomial_graph(20, 0.15)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.connected_caveman_graph(5, 10)
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```

```python
graph = nx.watts_strogatz_graph(500, 0, 0.05)
centrality = nx.closeness_centrality(graph)
colors = list(centrality.values())
plt.figure(figsize=(12, 6))
nx.draw(graph, 
        width=0.1,
        node_size=30,
        node_color=colors)
```

```python
plt.figure(figsize=(10, 15))
graph = nx.les_miserables_graph()
nx.draw(graph, 
        node_color='C1',
        with_labels=True)
```
